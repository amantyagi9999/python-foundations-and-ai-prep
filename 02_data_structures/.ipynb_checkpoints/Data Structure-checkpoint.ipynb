```python
# LISTS – ordered, mutable

# Create a list
numbers = [10, 20, 30, 40]

# Add elements
numbers.append(50)
print("After Append" , "::" , numbers)
numbers.insert(1, 15)
print("After insert" , "::" , numbers)

# Remove elements
numbers.remove(30)
print("After remove" , "::" , numbers)
popped = numbers.pop()  # removes last
print("After pop" , "::" , numbers)

# Access elements
first = numbers[0]
last = numbers[-1]
print("After first::" , first, "after last ::", last)

# Slicing
subset = numbers[1:3]
print("After slicing" , "::" , subset)

# Looping
for num in numbers:
    print(f"Number: {num}")

# List comprehension
squares = [x * 2 for x in numbers]
print("After Comprehesion" , "::" , squares)

print("Final list:", numbers)
print("Squares:", squares)
```

```python
# TUPLES – ordered, immutable

# Create a tuple
point = (10, 20, 30)

# Accessing values
x = point[0]
y = point[1]
print("Value x ::", x , "Value y::" , y)

# Tuple unpacking
a, b, c = point
print("After unpacking:",a,b,c,b)

# Tuple in dictionary key use case
location = {(10, 20): "Home"}

print("Point:", point)
print("Location value:", location.get((10, 20)))

```

```python
# SETS – unordered, unique elements

# Create set
items = {1, 2, 3, 4}

# Adding
items.add(5)

# Removing
items.discard(3)

# Check membership
print(2 in items)

# Set operations
a = {1, 2, 3}
b = {3, 4, 5}

print("Union:", a | b)
print("Intersection:", a & b)
print("Difference:", a - b)
print("Symmetric Difference:", a ^ b)
```

```python
# DICTIONARIES – key-value pairs

# Create dictionary
user = {
    "name": "Aman",
    "age": 30,
    "roles": ["ADMIN", "USER"]
}

# Accessing values
print(user["name"])
print(user.get("email", "Not Available"))

# Adding new key
user["email"] = "aman@example.com"

# Updating
user.update({"age": 31})

# Removing
user.pop("roles")

# Iterating
for key, value in user.items():
    print(key, ":", value)

print("Final dict:", user)
```

```python
# Pythonic transformations using comprehensions

# List comprehension
evens = [i for i in range(20) if i % 2 == 0]

# Dict comprehension
squares = {x: x*x for x in range(5)}

# Set comprehension
unique_chars = {c for c in "mifssissippi"}

print("Evens:", evens)
print("Squares:", squares)
print("Unique chars:", unique_chars)
```

```python
#Create a list of integers and:

#- Add 5 new values
number = [10, 40, 60, 90, 46]
print("Initial List::", number)

# - Remove duplicates
number.append(40)
print("Initial List after append::", number)

unique = list(set(number))
print("Unique::", unique)


#- Reverse the list without using reverse()

reversed_list = []

for i in numbers:
    reversed_list.insert(0, i)

print("reversed_list::", reversed_list)

#OR

reversed_list = numbers[::-1]
print(reversed_list)

```

```python
# Construct a dictionary from two lists: keys and values.

keys = ["name", "age", "city"]
values = ["Aman", 30, "Delhi"]

result = dict(zip(keys, values))
print(result)

# OR

result = {}
for i in range(len(keys)):
    result[keys[i]] = values[i]

print(result)

```

```python
# Write a function to return the frequency of each character using a dictionary.

def char_frequency(text):
    freq = {}

    for ch in text:
        freq[ch] = freq.get(ch,0) + 1
    return freq

print(char_frequency("mississippi"))
```

```python
# Given a tuple of numbers, return a new tuple with only even numbers.
nums = (1, 2, 3, 4, 5, 6)

def filter_even(nums):
    return tuple(n for n in nums if n % 2 == 0)

print(filter_even(nums))
```

```python
#  Convert a dictionary into list of tuples (key, value) using comprehension.

def dict_to_tuples(data):
    return [(k, v) for k, v in data.items()]

print(dict_to_tuples({"name": "Aman", "age": 30, "city": "Delhi"}))
```
